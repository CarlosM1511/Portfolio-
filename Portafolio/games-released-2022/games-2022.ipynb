```python
# Load Libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Read csv file
df = pd.read_csv("2022 Games.csv")
# Initial observatioon of the data
print(df.head())
print(df.dtypes)
```

```python
# Convert votes count column data type from string to float
df['votes count'] = df['votes count'].str.replace(',', '').astype(float)
# Drop Year column because all games are from 2022
df = df.drop('Year', axis=1)
print(df.head())
```

```python
print(df.head())
```

```python
# Fill votes count NaN values as 0
df['votes count'] = df['votes count'].fillna(0)
# brief description of votes count column 
print(df['votes count'].describe())
```

```python
top20 = df.nlargest(20, 'votes count')
sns.barplot(data=top20, x ='votes count', y ='Title')
plt.show()
```
