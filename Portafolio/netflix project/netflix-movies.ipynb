```python
# Import Libraries for data analysis

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read csv file and display first 5 rows

df = pd.read_csv("100 Best Movies on Netflix.csv")
print(df.head())
```

```python
# get info about columns and rows

print(df.info())
```

```python
print(df['Director'].unique())
```

```python
# List of directors that have more than one movie in the dataset
director_count = df.groupby("Director")["Movie Title"].count()
print(director_count[director_count > 1])
```

How does the distribution of movie scores look like in the top 100 Netflix movies dataset?

```python
# Histogram of Movies Score Distribution
plt.hist(df["Score"], bins=10)
plt.xlabel("Movie Score")
plt.ylabel("Frequency")
plt.title("Distribution of Movie Scores")
plt.show()
```

What is the trend in the number of movies produced over the years in the top 100 Netflix movies dataset?

```python
year_counts = df.groupby("Year")["Movie Title"].count()

# Line chart of movie production over the years
plt.plot(year_counts.index, year_counts.values)
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.title("Movie Production Over the Years")
plt.show()
```

Who are the most common cast members in the top 100 Netflix movies dataset?

```python
# Create a list of all cast members from the dataset
cast_list = []
for cast in df["Cast"]:
    cast_list.extend(cast.split(", "))

# Create a DataFrame of the most common cast members
cast_counts = pd.Series(cast_list).value_counts()
top_cast = cast_counts.head(10)

# Create a horizontal bar chart of the top cast members
plt.barh(top_cast.index, top_cast.values)
plt.xlabel("Frequency")
plt.ylabel("Cast Member")
plt.title("Most Common Cast Members in Top 100 Netflix Movies")
plt.show()
```
